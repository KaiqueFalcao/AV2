```python
import numpy as np

# Constantes
hbar = 1.055e-34  # Constante de Planck reduzida (J.s)
m = 9.11e-31      # Massa do elétron (kg)
eV = 1.602e-19    # 1 eV em Joules

# Energia da partícula e altura do potencial em Joules
E = 10 * eV
V = 9 * eV

# Cálculo dos vetores de onda
k1 = np.sqrt(2 * m * E) / hbar
k2 = np.sqrt(2 * m * (E - V)) / hbar

# Probabilidades de transmissão e reflexão
T = (4 * k1 * k2) / (k1 + k2)**2
R = ((k1 - k2) / (k1 + k2))**2

# Impressão dos resultados
print(f"Probabilidade de transmissão (T): {T:.4f}")
print(f"Probabilidade de reflexão (R): {R:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes do problema
g = 9.81  # aceleração da gravidade (m/s²)
y0 = 2.00  # altura inicial (m)
x_cesta = 10.0  # distância até a cesta (m)
y_cesta = 3.05  # altura da cesta (m)
angulo = np.radians(40)  # ângulo de lançamento em radianos
v0 = float(input("Digite a velocidade inicial (m/s): "))  # velocidade inicial (m/s)

# Componentes iniciais da velocidade
v0x = v0 * np.cos(angulo)
v0y = v0 * np.sin(angulo)

# Funções horárias
def x(t):
    return v0x * t
def y(t):
    return y0 + v0y * t - 0.5 * g * t**2
def vx(t):
    t = np.asarray(t)
    return v0x * np.ones_like(t)
def vy(t):
    return v0y - g * t
def ax(t):
    return 0
def ay(t):
    return -g

# a) Cálculo e impressão das posições x e y no tempo
t_total = 2 * v0y / g  # estimativa para tempo total (quando volta ao solo)
t = np.linspace(0, t_total, 300)

x_t = x(t)
y_t = y(t)

# b) Componentes de velocidade e aceleração
vx_t = vx(t)
vy_t = vy(t)
ax_t = ax(t)
ay_t = ay(t)

# c) Ponto de altura máxima (quando vy = 0)
t_max = v0y / g
x_max = x(t_max)
y_max = y(t_max)

# d) Velocidade e aceleração no ponto mais alto
vx_max = vx(t_max)
vy_max = vy(t_max)
ax_max = ax(t_max)
ay_max = ay(t_max)

# Impressões
print(f"Tempo até a altura máxima: {t_max:.2f} s")
print(f"Altura máxima: {y_max:.2f} m")
print(f"Posição x da altura máxima: {x_max:.2f} m")
print(f"Velocidade no ponto mais alto: vx = {vx_max:.2f} m/s, vy = {vy_max:.2f} m/s")
print(f"Aceleração no ponto mais alto: ax = {ax_max:.2f} m/s², ay = {ay_max:.2f} m/s²")

# Gráficos
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_t, y_t)
plt.title("Trajetória da bola")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.grid()
plt.axhline(y=y_cesta, color='red', linestyle='--', label='Altura da cesta')
plt.axvline(x=x_max, color='gray', linestyle='--', label='Altura máxima')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(t, vy_t, label='vy(t)')
plt.plot(t, vx_t, label='vx(t)')
plt.title("Componentes da velocidade")
plt.xlabel("Tempo (s)")
plt.ylabel("Velocidade (m/s)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Entrada de dados
r = float(input("Digite o valor de r (distância): "))
theta_graus = float(input("Digite o valor de θ (em graus): "))

# Conversão de graus para radianos
theta = np.radians(theta_graus)

# Conversão para coordenadas cartesianas
x = r * np.cos(theta)
y = r * np.sin(theta)

# Saída dos resultados
print(f"As coordenadas cartesianas são: x = {x:.2f}, y = {y:.2f}")
```

```python
# Leitura do valor de N
N = int(input("Digite um valor inteiro positivo para N: "))

# Cálculo da soma
S = 0
for i in range(1, N + 1):
    S += i / (N - i + 1)

# Exibe o resultado
print(f"O valor de S é: {S:.2f}")
```
